```python
#dropping happiness rank beacuse it depends on happiness score

df.drop(labels=['Happiness Rank'],axis=1,inplace=True)
```

```python
x=df.iloc[:,1:]
x.head()
```

```python
x.shape
```

```python
y=df.iloc[:,0:1]
y.head()
```

```python
y.shape
```

```python
#model building

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, mean_absolute_error
from sklearn.metrics import r2_score
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.25,random_state=45)
print('xtrain shape--->',x_train.shape)
print('ytrain shape--->',y_train.shape)
print('xtest shape--->',x_test.shape)
print('ytest shape--->',y_test.shape)
```

```python
#we use linear regression model because it has all continuous variables

lrm=LinearRegression()
lrm.fit(x_train,y_train)
print(lrm.coef_)
print(lrm.intercept_)
```

```python
score=lrm.score(x_train,y_train)
score
```

```python
pred=lrm.predict(x_test)
print('predicted result:',pred)
print('actual result:',y_test)
```

```python
print('errors:')
print('mean squared error:',mean_squared_error(y_test,pred))
print('mean absolute error:',mean_absolute_error(y_test,pred))
print('root mean squared error:',np.sqrt(mean_squared_error(y_test,pred)))
print('r2 score:',r2_score(y_test,pred))
print('score:',score*100,'%')
```

interpretation:
    
    as we see that we used linear model and our score is 99.99%
    for further prediction we use linear model.
